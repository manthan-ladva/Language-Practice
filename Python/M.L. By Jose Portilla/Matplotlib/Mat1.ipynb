```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
plt.plot(x,y)
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Title")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'green')

plt.subplot(1,2,2)
plt.plot(y,x, 'violet')
```

#OO

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('OO Title')
```

```python
fig= plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1.set_title('1st Plot')

axes2 = fig.add_axes([0.4,0.3,0.4,0.3])
axes2.set_title('2nd Plot')

axes3 = fig.add_axes([0.3,0.15,0.4,0.3])
axes3.set_title('3rd Plot')

axes1.plot(x,y)
axes2.plot(y,x)
axes3.plot(x,y)
```

#Lecture -- 2 %%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
plt.tight_layout()
for cur_ax in axes:
    cur_ax.plot(x,y,'orange')
```

```python
fi = plt.figure(figsize=(8,2))
ax = fi.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2, ncols=1, figsize=(8,4))
#ax = fig.add_axes([0,0,1,1])
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig("myfig.png")
```

```python
fi = plt.figure()
ax = fi.add_axes([0,0,1,1])

ax.plot(x,x**2, label = "X Square")
ax.plot(x,x**3, label = 'X Cube')

ax.legend(loc = 10)
```
