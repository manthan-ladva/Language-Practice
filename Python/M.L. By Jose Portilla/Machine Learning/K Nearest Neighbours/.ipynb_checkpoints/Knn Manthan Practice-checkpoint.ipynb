```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv("Classified Data", index_col=0)
data.head(10)
```

```python
data.info()
```

```python
data.describe()
```

```python
from sklearn.preprocessing import StandardScaler

stdsclar = StandardScaler()
```

```python
scaled_feature = stdsclar.fit_transform(data.drop('TARGET CLASS', axis=1))
scaled_feature
```

```python
data_scal = pd.DataFrame(scaled_feature, columns=data.columns[:-1])
data_scal.head(10)
```

```python
from sklearn.model_selection import train_test_split

X = data_scal
y = data['TARGET CLASS']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn= KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train, y_train)
```

```python
pred = knn.predict(X_test)
pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(confusion_matrix(y_test, pred))
print(classification_report(y_test, pred))
```

```python
error_rate = []

for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i !=y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40), error_rate, color='red', linestyle='--', marker='o', markerfacecolor='pink', markersize=10)
```

```python
knn= KNeighborsClassifier(n_neighbors=38)
knn.fit(X_train, y_train)
pred = knn.predict(X_test)

print(confusion_matrix(y_test, pred))
print('\n')
print(classification_report(y_test, pred))
```
