```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

```python
g = sns.PairGrid(iris)                              #--------------Pair Grid plot wil give blank grid maps
g.map_diag(sns.distplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.FacetGrid(data=tips, row='smoker', col='time').map(sns.distplot, 'total_bill')
```

```python
sns.FacetGrid(data=tips, row='smoker', col='time').map(sns.scatterplot, 'total_bill', 'tip')
```
