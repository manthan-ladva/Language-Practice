```python
import seaborn as sns
%matplotlib inline

tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o','v'],     ### where, hue = category
           scatter_kws = {'s':70})
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, row='sex', col='time')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', hue='sex', aspect=0.6, height=8)
```
